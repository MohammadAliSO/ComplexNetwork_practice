```python
import networkit as nk
import networkx as nx
import matplotlib.pyplot as plt
```

ZACHARY KARATE CLUB

```python

G_nx = nx.karate_club_graph()

G = nk.Graph(n=len(G_nx), weighted=False, directed=False)
for edge in G_nx.edges():
    G.addEdge(edge[0], edge[1])

print('#nodes:', len(G_nx.nodes()), 'and', '#edges:', len(G_nx.edges()))
```

```python
pos = nx.spring_layout(G_nx)
nx.draw(G_nx, pos, with_labels=True, node_color='skyblue', node_size=300)
plt.title('Zachary Karate Club Network')
plt.show()
```

```python
#k_path Centrality
k = 4  # The desired path length
k_path_centrality = nk.centrality.KPathCentrality(G, alpha=0.2, k=0).run().scores()
print(k_path_centrality)
```

```python

for node, centrality in enumerate(k_path_centrality):
    G_nx.nodes[node]['kpath_centrality'] = centrality

#Normilize Centrality
centrality_values = list(nx.get_node_attributes(G_nx, 'kpath_centrality').values())
normalized_centrality = [(value - min(centrality_values)) / (max(centrality_values) - min(centrality_values)) for value in centrality_values]

node_size = [v * 300 for v in normalized_centrality]

pos = nx.spring_layout(G_nx)
nx.draw(G_nx, pos, with_labels=True, node_size=node_size,node_color='skyblue', cmap=plt.cm.RdYlBu, font_size=10)
plt.title('Zachary Karate Club Network (Node Sizes based on K-Path Centrality)')
plt.show()
```
