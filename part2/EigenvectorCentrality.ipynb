```python
import networkx as nx
import matplotlib.pyplot as plt
```

ZACHARY KARATE CLUB

```python

G = nx.karate_club_graph()
print('#nodes:', len(G.nodes()), 'and', '#edges:', len(G.edges()))
```

```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=300)
plt.title('Zachary Karate Club Network')
plt.show()
```

```python
#Eigenvector Centrality
eigenvector_centrality = nx.eigenvector_centrality(G)

print(eigenvector_centrality)
```

```python
node_sizes = [3000 * centrality for centrality in eigenvector_centrality.values()]
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=node_sizes)
plt.title('Zachary Karate Club Network (Node Sizes based on Degree Centrality)')
plt.show()
```

Dolphins Network

```python
G = nx.read_edgelist("out.dolphins", create_using = nx.Graph(), nodetype = int)
print('#nodes:', len(G.nodes()), 'and', '#edges:', len(G.edges()))
```

```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='orange', node_size=200 )

plt.title('Dolphins Network')
plt.show()
```

```python
#Eigenvector Centrality
eigenvector_centrality = nx.eigenvector_centrality(G)

print(eigenvector_centrality)
```

```python
node_sizes = [3000 * centrality for centrality in eigenvector_centrality.values()]
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='orange', node_size=node_sizes)
plt.title('Dolphins Network (Node Sizes based on Eigenvector  Centrality)')
plt.show()
```
