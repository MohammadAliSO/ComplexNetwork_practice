```python
import numpy as np
import skfuzzy as fuzz
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset = nx.karate_club_graph()
G = nx.Graph(dataset)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=200, node_color='skyblue', font_size=10)
plt.title('Zachary Karate Club Network Graph')
plt.show()
```

```python
def run_fcm(matrix, n_clusters):
    cntr, u, _, _, _, _, _ = fuzz.cluster.cmeans(
        matrix, n_clusters, 2, error=0.005, maxiter=1000)
    return cntr, u
```

```python
adjacency_matrix = nx.to_numpy_array(G, dtype=int)
df_adjacency_matrix = pd.DataFrame(adjacency_matrix)
df_adjacency_matrix
```

```python
#cluster count
n_clusters = 2

#FCM
cntr, u = run_fcm(adjacency_matrix, n_clusters)
df_U = pd.DataFrame(u.T)
df_U
```

```python
sns.clustermap(u.T)
```

```python
cluster_membership = np.argmax(u, axis=0)
cluster_membership
```

```python

colors = plt.cm.rainbow(np.linspace(0, 1, n_clusters))

pos = nx.spring_layout(G)
for i in range(n_clusters):
    nodes = [node for node in G.nodes() if cluster_membership[node] == i]
    nx.draw_networkx_nodes(G, pos, nodelist=nodes, node_color=colors[i], node_size=300)

nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)
plt.title('Zachary Network Graph Cluster')
plt.show()
```
