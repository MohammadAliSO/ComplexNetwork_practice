```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.decomposition import NMF
import seaborn as sns
```

```python
dataset = nx.karate_club_graph()
G = nx.Graph(dataset)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=200, node_color='skyblue', font_size=10)
plt.title('Jazz Network Graph')
plt.show()
```

```python
def run_nmf(matrix, n_components):
    model = NMF(n_components=n_components, init='random', random_state=0)
    w = model.fit_transform(matrix)
    h = model.components_
    return model,w,h
```

```python
adjacency_matrix = nx.to_numpy_array(G, dtype=int)
df_adjacency_matrix = pd.DataFrame(adjacency_matrix)
df_adjacency_matrix
```

```python
#component count
n_components = 2

#NMF
model,W,H = run_nmf(adjacency_matrix, n_components)

df_W = pd.DataFrame(W)
df_W
```

```python
norms = np.linalg.norm(W.T, ord=2, axis=0)

normalized_W = W.T / norms
W =  normalized_W.T

df_W=pd.DataFrame(W)
df_W
```

```python
sns.clustermap(W)
```

```python
cluster_membership = np.argmax(W, axis=1)
colors = plt.cm.rainbow(np.linspace(0, 1, n_components))

pos = nx.spring_layout(G)
for i in range(n_components):
    nodes = [node for node in G.nodes() if cluster_membership[node] == i]
    nx.draw_networkx_nodes(G, pos, nodelist=nodes, node_color=colors[i], node_size=300)

nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)
plt.title('Zachary Network Graph Cluster')
plt.show()
```
